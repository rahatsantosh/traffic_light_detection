```python
import cv2
import os
import numpy as np

f = './dataset'
x={}
folder=os.listdir(f)
for filename in folder:
    i=0
    full_path = os.path.join(f, filename)
    y=[]
    if os.path.isdir(full_path):
        for fg in os.listdir(full_path):
            im = os.path.join(full_path, fg)
            if im.endswith(".jpg") or im.endswith(".jpeg"):
                imgs=cv2.imread(im,cv2.IMREAD_GRAYSCALE)
                new=cv2.resize(imgs,(200,200))
                y.append([new])
        x[filename]=y
print(len(x))
print(np.shape(x['light_1']))
```

```python
# label -> (left,right,straight,light_0,light_1,u)
data=[]
for (name, y) in x.items():
    if name == 'light_0':
        label = 1
    elif name == 'light_1':
        label = 2
    else:
        label = 0
    for a in y:
        data.append([a,label])
print(len(data))
```

```python
import random
random.shuffle(data)
```

```python
x=[]
y=[]
for a in data:
    x.append(a[0])
    y.append(a[1])
print(x)
print(np.shape(x))
print(y)
print(np.shape(y))
```

```python
import pickle
pickle_out=open('data.pickle','wb')
pickle.dump(x,pickle_out)
pickle_out.close()

pickle_out=open('label.pickle','wb')
pickle.dump(y,pickle_out)
pickle_out.close()
```
