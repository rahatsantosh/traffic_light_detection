```python
def segment(img,model,segment_size):
    assert np.shape(img)[2]==1
    n, m = segment_size
    x, y, _ = np.shape(img)
    predict = np.empty((x//n,y//m), dtype=int)
    for i in range(x//n):
        for j in range(y//m):
            new = img[i*n:(i+1)*n, j*m:(j+1)*m]
            new = np.resize(new, (1, 200, 200, 1))
            res = model.predict(new)
            predict[i, j] = np.argmax(res)
    return predict
```
